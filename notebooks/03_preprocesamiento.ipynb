# 03 - Preprocesamiento

**Objetivo:** dejar los datos listos para que un modelo los pueda usar:

1. Separar las **X** (características) de la **y** (target `smoking`).
2. Sacar columnas que no sirven (`ID`, `oral`).
3. Convertir las columnas de **texto a número**.
4. **Partir** los datos en entrenamiento y validación.
5. Guardar los datos procesados en `data/processed/`.



```python
# Imports
import pandas as pd
import numpy as np

# Cargo los datos etiquetados
train = pd.read_excel("../data/raw/smoking_train_labeled.xlsx")

print("Datos cargados:", train.shape)
train.head(3)
```

```python
# y = el target (lo que queremos predecir)
y = train["smoking"]

# X = las características: TODO menos el target y las columnas inútiles
X = train.drop(columns=["smoking", "ID", "oral"])

print("X (características):", X.shape)
print("y (target):       ", y.shape)
X.head(3)
```

y = solo la respuesta (la columna smoking)
X = solo las pistas (las demás columnas), tirando 3 que no servían (smoking, ID, oral)

```python
# Convierto las columnas de texto a número, encoding
X["gender"] = X["gender"].map({"F": 0, "M": 1})
X["tartar"] = X["tartar"].map({"N": 0, "Y": 1})

# Verifico cómo quedaron
print(X[["gender", "tartar"]].dtypes)
X[["gender", "tartar"]].head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X, y,
    test_size=0.2,      # 20% para validación, 80% para entrenar
    random_state=42,    # fija el azar: el corte es siempre igual (reproducible)
    stratify=y          # mantiene la proporción de fumadores en ambas partes
)

print("Entrenamiento:", X_train.shape, "  y_train:", y_train.shape)
print("Validación:   ", X_val.shape, "  y_val:", y_val.shape)
```

```python
from pathlib import Path

PROC = Path("../data/processed")   # carpeta donde guardo

# Guardo las 4 piezas como archivos CSV
X_train.to_csv(PROC / "X_train.csv", index=False)
X_val.to_csv(PROC / "X_val.csv", index=False)
y_train.to_csv(PROC / "y_train.csv", index=False)
y_val.to_csv(PROC / "y_val.csv", index=False)

print("¡Guardado en data/processed/! ✅")

import os
print("Archivos:", os.listdir(PROC))
```
