# 06 - Predicción (Entrega final)

**Objetivo:** generar las predicciones sobre el conjunto **sin etiquetar** (5.692 personas).

Pasos:
1. Cargar los datos sin etiquetar.
2. Aplicar **el mismo preprocesamiento** que usamos al entrenar.
3. Cargar el modelo entrenado y predecir (con el umbral 0,39).
4. Agregar la columna `smoking_prediction` (0/1) y exportar el archivo.



```python
import pandas as pd
import joblib

# Cargo los datos SIN etiquetar (los 5.692 que hay que predecir)
test = pd.read_excel("../data/raw/smoking_unlabeled.xlsx")

# Cargo el modelo entrenado y el umbral
modelo = joblib.load("../models/random_forest.joblib")
umbral = joblib.load("../models/umbral.joblib")

print("Datos a predecir:", test.shape)
print("Modelo y umbral cargados ✅  | Umbral:", umbral)
```

```python

# Aplico EL MISMO preprocesamiento que en el entrenamiento

# 1) Saco las columnas que no usamos (acá NO hay 'smoking', solo ID y oral)
X_test = test.drop(columns=["ID", "oral"])

# 2) Convierto el texto a número, con el mismo mapeo de antes
X_test["gender"] = X_test["gender"].map({"F": 0, "M": 1})
X_test["tartar"] = X_test["tartar"].map({"N": 0, "Y": 1})

# 3) Aseguro que las columnas estén en el MISMO orden que al entrenar
X_test = X_test[modelo.feature_names_in_]

print("X_test listo:", X_test.shape)
X_test.head(3)
```

```python
# Exporto el dataset completo CON la columna de predicciones
salida = "../data/processed/smoking_prediction_entrega.csv"
test.to_csv(salida, index=False)

print("✅ Archivo de entrega guardado en:", salida)
print("Shape:", test.shape)
print("¿Está la columna smoking_prediction?:", "smoking_prediction" in test.columns)
```

```python
probs_test = modelo.predict_proba(X_test)[:, 1]
predicciones = (probs_test >= umbral).astype(int)
test["smoking_prediction"] = predicciones

print("Predicciones generadas:", len(predicciones))
print(test["smoking_prediction"].value_counts())
```

```python
salida = "../data/processed/smoking_prediction_entrega.csv"
test.to_csv(salida, index=False)

print("✅ Guardado:", salida)
print("Shape:", test.shape)
print("¿Está smoking_prediction?:", "smoking_prediction" in test.columns)
```
