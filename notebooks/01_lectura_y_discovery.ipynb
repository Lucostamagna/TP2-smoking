## Conclusiones del Discovery

- El dataset de entrenamiento (`train`) tiene **50.000 filas y 27 columnas**. El de predicción (`test`) tiene **5.692 filas y 26 columnas**: le falta `smoking`, que es justamente el **target** que vamos a predecir.
- **No hay datos faltantes ni filas duplicadas** → los datos están limpios.
- El target `smoking` está **desbalanceado**: ~63% no fuma (0) y ~37% fuma (1).`M`
- Hay **3 columnas de texto**: `gender` (M/F), `oral` y `tartar` (Y/N). Habrá que convertirlas a número.
- **`oral` es constante** (siempre 'Y') → se descarta porque no aporta información.
- **`ID`** es solo un identificador → se descarta para entrenar (no es una característica real).
- Las columnas numéricas están en **escalas muy distintas** → habrá que escalarlas para modelos como KNN.
- Algunas columnas de hígado (`AST`, `ALT`, `Gtp`) tienen **valores extremos (outliers)**, algo normal en datos médicos.
