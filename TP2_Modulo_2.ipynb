# Trabajo Práctico 2 - Módulo 2

Llegamos al final del módulo 2 y estamos en condiciones de analizar un dataset, entrenar un modelo de clasificación y evaluar sus resultados.

## Objetivo del Proyecto

El objetivo de este proyecto es construir un modelo de Machine Learning capaz de predecir si una persona es fumadora o no (`smoking`), a partir de un conjunto de características.

Para lograrlo, deberán seguir un flujo de trabajo de ciencia de datos completo y profesional:
1. **Entrenamiento y Validación:** Utilizarán un primer conjunto de datos etiquetados para entrenar y validar su modelo.
2. **Predicción sobre datos nuevos:** Utilizarán el modelo entrenado y **el mismo pipeline de preparación de datos** para procesar un segundo conjunto de datos (sin etiquetas) y generar predicciones.
3. **Formato de la predicción:** Las predicciones sobre el conjunto sin etiquetas deberán enviarse en una nueva columna llamada `"smoking_prediction"` con los valores directamente como `0` o `1`.

## Formato de Entrega y Requisitos

La entrega del proyecto debe simular un entorno de trabajo real, por lo que se exigirá un repositorio público en **GitHub** con una estructura ordenada y reproducible.

### 1. Repositorio de GitHub
Deberán entregar el enlace a un repositorio público que contenga todo el proyecto.

### 2. Estructura del Proyecto
El repositorio debe estar organizado en carpetas de la siguiente manera:

*   `data/`: Contendrá los datos del proyecto. Debe dividirse en:
    *   `raw/`: Datos originales, tal como se descargaron.
    *   `processed/`: Datos limpios y procesados listos para modelar.
    *   `external/`: Cualquier otro dato externo utilizado.
*   `models/`: Modelos entrenados y guardados (ej. archivos `.pkl, .joblib`).
*   `notebooks/`: Aquí irán las notebooks de Jupyter, ordenadas lógicamente, por ejemplo:
    *   `01_lectura_y_discovery.ipynb`
    *   `02_eda.ipynb` (Análisis Exploratorio de Datos)
    *   `03_preprocesamiento.ipynb` (Limpieza, feature engineering)
    *   `04_entrenamiento_y_optimizacion.ipynb`
    *   `05_validacion.ipynb`
    *   `06_prediccion.ipynb` (Debe contener el código que genera las predicciones finales)
*   **Archivos adicionales requeridos:** Script de requerimientos (ej. `requirements.txt`).

**Importante:** Todo el código debe poder ejecutarse y reproducirse sin errores.

### 3. README.md Completo
El repositorio debe incluir un archivo `README.md` claro y detallado que documente:
*   Descripción del proyecto y objetivo.
*   Instrucciones para reproducir el entorno y ejecutar el código.
*   Explicación de la estructura de archivos.
*   Resumen de los experimentos realizados, pruebas, intentos y cambios tomados a lo largo del desarrollo.
*   Conclusiones principales.

### 4. Generación de las Predicciones
En una de las notebooks (ej. `06_prediccion.ipynb`), debe estar claramente visible el código que:
1. Carga los datos sin etiquetar.
2. Aplica las mismas transformaciones (pipeline) usadas en el entrenamiento.
3. Genera las predicciones con el modelo elegido.
4. Exporta el archivo resultante con la nueva columna `"smoking_prediction"`.

### 5. Métrica de Éxito
La métrica para evaluar el rendimiento de su modelo será el **F1-Score** para target = 1.
El valor de performance definitivo se calculará sobre las predicciones que envíen del segundo conjunto de datos (sin etiquetas), y se revelará posteriormente a las presentaciones de los trabajos.
